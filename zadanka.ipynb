##### Aktywność źródła  - A
Aktywność, zdefiniowana już w poprzednim rozdziale, jest liczbą rozpadów
promieniotwórczych w źródle, zachodzących w jednostce czasu. Jednostką
aktywności jest bekerel (od Henri Becquerela – odkrywcy promieniotwórczości
naturalnej):
    
$$1 \text{Bq} = 1  \text{rozpad/s}.$$

##### Aktywność właściwa - a
Jest to aktywność na jednostkę masy.

#### 1. Po jakim czasie aktywność 1 kg preparatu zawierającego 250g $K^{131}I$ o aktywności właściwej 16 MBq/kg zmniejszy się do poziomu 1 MBq, wiedząc, że T_1/2 $^{131}I$ wynosi 8 dni?   

```python
#Aktywność całkowita

#Ac=A0+0 
#widzimy więc że interesuje nas tylko masa K 131-I


#masa preparatu wyrażona w kilogramach
m = 0.25          #kg 

#aktywność promieniotwórcza właściwa wyrażona w MBq/kg (specific activity)
a0 = 16  #MBq/kg

#aktywność promieniotwórcza - A wyrażona w Becquerelach (activity)
A0 = m * a0         #MBq

#Wiemy, że aktywność promieniotwórcza zależy od stałej rozpadu promieniotwórczego i czasu według wzoru:

#   A=A_0*Exp[-lamb*t],         gdzie lambda=ln(2)/T_12, T_12 to czas połowicznego rozpadu.

T_12 = 8 * 24 * 3600  #sekundy 
lamb = np.log(2)/T_12 #stała rozpadu

t = np.linspace(0,T_12*10) #czas

A = A0*np.exp(-lamb * t)
```

```python
#Wyrysuj funkcję aktywności of czasu A(t):

plt.ylabel("A(t)[MBq]",fontsize=18)
plt.xlabel("t[s]", fontsize=18)
plt.plot(t,np.ones(np.size(t)))

diff=np.abs(1.0-A)
smallest_index = np.argmin(diff)

plt.plot(t, A)
plt.plot(t[smallest_index],A[smallest_index],"o")
plt.show()

print("Aktywność spadnie do 1MBq po czasie dwoch polowicznych czasow rozpadu: t = {:} [1/T_12]!"\
      .format(t[smallest_index]/(3600*8*24)))
```

#### 2. Oblicz jaką masę będzie miał po 32 dniach 1 kg preparatu zawierający 32 kBq 131J, wiedząc, że T1/2 131-I wynosi 8 dni.

```python
#mu*N(t)=mu*N(0)e^{-lamb*t}, mu => masa cząstki preparatu

m0 = 1     #kilogram
T_12 = 8   #dni

#Obliczmy A:
t = np.linspace(0,4*T_12) #32 dni
lamb = np.log(2)/T_12 #stała rozpadu

m=m0*np.exp(-lamb*t)
print("Po 32 dniach z preparatu o masie {} kg pozostanie {} g.".format(m0,m[-1:][0]*1000))
```

```python
#Wyrysuj funkcję aktywności of czasu A(t):

plt.ylabel("m(t) [kg]",fontsize=18)
plt.xlabel("t [dni]", fontsize=18)

plt.plot(t, m)
plt.show()
```

#### 4. Jaką aktywność właściwą będzie miał po 30 latach 2 kg preparatu zawierającego 32 GBq 137Cs? (T1/2 137Cs = 30 lat)?

```python
m0 = 2   #kilogram
A0 = 32 #GBq
T_12 = 30  #lat


t = np.linspace(0,1*T_12) #sekund
lamb = np.log(2)/T_12 #stała rozpadu

#Obliczamy A
A=A0*np.exp(-lamb*t)

#Obliczmy m:
m=m0*np.exp(-lamb*t)

print("Po 30 latach aktywność wyniesie {} GBq".format(A[-1:][0]))
print("Po 30 latach z preparatu o masie {} kg pozostanie {} kg.".format(m0,m[-1:][0]))
print("Po 30 latach aktywność właściwa wyniesie {} GBq/kg".format(A[-1:][0]/m[-1:][0]))
print("WNIOSEK aktywność właściwa jest stała!")
```

```python
plt.ylabel("a(t)= A(t)m(t) [GBq/kg]",fontsize=18)
plt.xlabel("t [lat]", fontsize=18)

plt.plot(t,A/m)
plt.show()
```

#### 11. Wiedząc, że dla chloru Lct100 = 100 mg . min/ m3 , a LD100 = 0,06 mg/kg, oblicz minimalną ilość gazu, która wprowadzona poprzez drogi oddechowe do organizmu mężczyzny ważącego 75 kg spowoduje jego śmierć. Proszę przyjąć, jeśli to konieczne, że człowiek oddycha średnio 16 razy na minutę, a podczas jednego oddechu do organizmu wprowadza 0,5 dm3 powietrza.

```python
#sprawdzić haber's rule

Lct_100=100 #mg*min/m^3
LD_100=0.06 #mg/kg

breathing_rate=16*0.5*pow((1/10),3) #razy/min*m^3

masa_mezczyzny=75 #kg

wt=0.12  #dla płóc

#wt*(Lct_100*breathing_rate/masa_mezczyzny)*time=LD_100

time=wt*(Lct_100*breathing_rate/masa_mezczyzny)/LD_100
print("czas w minutach po ktorym mężczyzna się otruje: {:.2} min.".format(time))

print("Szukana masa gazu potrzebna do otrucia mężczyzny wynosi: {:.2} kg.".format(Lct_100*breathing_rate*time))
```
